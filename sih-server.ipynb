```python
import os
import multiprocessing
import torch
import chromadb
import fitz
from chromadb.config import Settings

from langchain.llms import LlamaCpp
from langchain.chains.summarize import load_summarize_chain
from langchain.prompts import PromptTemplate
from langchain.text_splitter import TokenTextSplitter
from langchain.docstore.document import Document
from langchain.chains import LLMChain
from langchain.embeddings import GPT4AllEmbeddings, HuggingFaceEmbeddings
from langchain.chains.question_answering import load_qa_chain
from langchain.memory import ConversationBufferMemory
from langchain.vectorstores import Chroma

from langchain.text_splitter import TokenTextSplitter, RecursiveCharacterTextSplitter
from langchain.retrievers import ParentDocumentRetriever
from langchain.storage import InMemoryStore
from pprint import pprint
from transformers import LlamaTokenizerFast
from langchain.chains import AnalyzeDocumentChain

from langchain.chains.combine_documents.stuff import StuffDocumentsChain
from langchain.chains import MapReduceDocumentsChain, ReduceDocumentsChain
from langchain.text_splitter import CharacterTextSplitter

import spacy
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import torch
import torch.nn as nn

# Load pre-trained spaCy model
nlp = spacy.load("en_core_web_lg")
cos_sim = nn.CosineSimilarity(dim=1, eps=1e-6)

# MODEL_PATH = "orca_mini_v3_7b.Q8_0.gguf"
MODEL_PATH = "llama-2-7b-chat.Q8_0.gguf"
llm = LlamaCpp(model_path=MODEL_PATH, n_ctx=4096, n_gpu_layers=-1, max_tokens=4096, n_batch=3900, temperature=0)
# Embedding model initialization
embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2",
    model_kwargs={'device': 'cuda'},
    encode_kwargs={'normalize_embeddings': False}
)
```

```python
# Chroma DB
persist_directory = 'db'
CHROMA_SETTINGS = Settings(persist_directory=persist_directory, anonymized_telemetry=False)
chroma_client = chromadb.PersistentClient(settings=CHROMA_SETTINGS, path=persist_directory)

# Prompts
SUMMARY_PROMPT_TEMPLATE = """
### System:
You are an AI assistant. You will be given a task. You must generate a detailed and long answer.

### User:
Summarize the following text.
{text}

### Response:
Sure, here is a summary of the text:
"""

NEW_SUMMARY_PROMPT_TEMPLATE_50 = """
System: You are a summarization assistant. You will be given a task and are expected to generate a brief answer.
User:
Summarize the following text in no more or no less than 50 words.
{text}
Assistant:
"""

NEW_SUMMARY_PROMPT_TEMPLATE_100 = """
System: You are a summarization assistant. You will be given a task and are expected to generate a detailed answer.
User:
Summarize the following text.
{text}
Assistant:
"""

GRAMMAR_PROMPT_TEMPLATE = """
### System:
You are an AI assistant that follows instruction extremely well. Help as much as you can.

### User:
Read the following text, and rewrite all sentences after correcting all the writing mistakes:
{text}

### Response:
"""

NEW_GRAMMAR_PROMPT_TEMPLATE = """
System: You are an AI assistant that follows instruction extremely well. Help as much as you can.
User:
Read the following text, and rewrite all sentences after correcting all the writing mistakes. Don't add anything else, other than the corrected text.
{text}
Assistant:
This is the corrected text:
"""

CHAT_PROMPT_TEMPLATE = """
### System:
You are an AI assistant that helps people find information.

### User:
This is your previous chat history, where "Human:" is the user's query and "AI:" is your response to the query:
{chat_history}

This is the information provided to you:
{context}

Use only the conversation history (if there was previous conversations made) and the information to answer the following query.
{question}

### Response:
"""

# # Map
# map_template = """
# ### User:
# Summarize the following text, shrink the given information below if and only if its big. Try to retain core information as much as possible:
# {docs}

# ### Response:
# Sure, here is a summary of the text:
# """

# # Reduce
# reduce_template = """
# ### User:
# Summarize the following text, shrink the given information below if and only if its big. Try to retain core information as much as possible:
# {docs}

# ### Response:
# Sure, here is a summary of the text:
# """

# map_prompt = PromptTemplate.from_template(map_template)
# reduce_prompt = PromptTemplate.from_template(reduce_template)

summarize_prompt_50 = PromptTemplate.from_template(NEW_SUMMARY_PROMPT_TEMPLATE_50)
summarize_prompt_100 = PromptTemplate.from_template(NEW_SUMMARY_PROMPT_TEMPLATE_100)
grammar_prompt = PromptTemplate(template=NEW_GRAMMAR_PROMPT_TEMPLATE, input_variables=["text"])
qa_prompt = PromptTemplate(template=CHAT_PROMPT_TEMPLATE, input_variables=["question", "chat_history", "context"])

def pdf_to_ocr(path):
    doc = fitz.open(path)
    content = ''
    for page in doc:
        content += page.get_text()
    return content

def summarize_pdf(text, no_of_words):
    # Convert pdf to text
    # text = pdf_to_ocr(pdf_path)
    # map_chain = LLMChain(llm=llm, prompt=map_prompt)
    # reduce_chain = LLMChain(llm=llm, prompt=reduce_prompt)

    # # Takes a list of documents, combines them into a single string, and passes this to an LLMChain
    # combine_documents_chain = StuffDocumentsChain(
    #     llm_chain=reduce_chain, document_variable_name="docs"
    # )

    # # Combines and iteravely reduces the mapped documents
    # reduce_documents_chain = ReduceDocumentsChain(
    #     # This is final chain that is called.
    #     combine_documents_chain=combine_documents_chain,
    #     # If documents exceed context for `StuffDocumentsChain`
    #     collapse_documents_chain=combine_documents_chain,
    #     # The maximum number of tokens to group documents into.
    #     token_max=4000,
    # )

    # # Combining documents by mapping a chain over them, then combining results
    # map_reduce_chain = MapReduceDocumentsChain(
    #     # Map chain
    #     llm_chain=map_chain,
    #     # Reduce chain
    #     reduce_documents_chain=reduce_documents_chain,
    #     # The variable name in the llm_chain to put the documents in
    #     document_variable_name="docs",
    #     # Return the results of the map steps in the output
    #     return_intermediate_steps=False,
    # )

    # Tokenize and get document embedding
    document = text
    doc_embedding = np.mean([word.vector for word in nlp(document)], axis=0)

    # Tokenize and get sentence embeddings
    sentences = [sent.text for sent in nlp(document).sents]
    sentence_embeddings = [np.mean([word.vector for word in nlp(sent)], axis=0) for sent in sentences]


    sentence_embeddings = torch.FloatTensor(sentence_embeddings).cuda()
    doc_embedding = torch.FloatTensor(doc_embedding).cuda()

    similarity_scores = cos_sim(sentence_embeddings, doc_embedding.view(1,-1)).cpu().numpy().tolist()

    # Rank sentences based on similarity scores
    sentences = [(i, j) for i,j in enumerate(sentences)]
    ranked_sentences = sorted(list(zip(sentences, similarity_scores)), key=lambda x: x[1], reverse=True)

    limit = 3700 #safety
    curr = 0
    iter_ = 0
    total_sent = len(ranked_sentences)

    selected = []
    tokenizer = LlamaTokenizerFast.from_pretrained("hf-internal-testing/llama-tokenizer")

    while curr < limit and iter_ < total_sent:
      selected.append(ranked_sentences[iter_][0])
      iter_+=1

      curr += len(tokenizer.encode(ranked_sentences[iter_][0][1]))*1.2 #safety
    ranked_sentences = sorted(selected, key=lambda x: x[0], reverse=False)
    text = ''
    for i in ranked_sentences:
      text += i[1]
    print (text)
    docs = [Document(page_content=text)]

    # # Split text into chunks
    # text_splitter = TokenTextSplitter(chunk_size=2000)
    # texts = text_splitter.split_text(text)
    # docs = [Document(page_content=t) for t in texts]

    summary_chain = load_summarize_chain(llm, chain_type='stuff', prompt=summarize_prompt_100)

    # if (no_of_words == 50):
    #   summary_chain = load_summarize_chain(llm, chain_type='map_reduce', map_prompt=summarize_prompt_50, combine_prompt=summarize_prompt_50)
    # elif (no_of_words == 100):
    #   summary_chain = load_summarize_chain(llm, chain_type='map_reduce', map_prompt=summarize_prompt_100, combine_prompt=summarize_prompt_100)

    # print (summary_chain, docs)
    # Run inference
    try:
        result = summary_chain.run(docs)
    except Exception as e:
        return e

    return result

def grammar_check(text):
    llm_chain = LLMChain(prompt=grammar_prompt, llm=llm)
    # Run inference
    try:
        result = llm_chain.run(text)
        return result

    except Exception as e:
        return e

def ingest_file(text):
    # Convert pdf to text
    try:
        # text = pdf_to_ocr(pdf_path)
        # print (text)
        # text_splitter = TokenTextSplitter(chunk_size=200, chunk_overlap=0)
        # texts = text_splitter.split_text(text)
        # docs = [Document(page_content=t) for t in texts]
        # db = Chroma.from_documents(docs, embeddings,
        #                            persist_directory=persist_directory,
        #                            client=chroma_client,
        #                            client_settings=CHROMA_SETTINGS)
        chunk_size = 300
        chunk_overlap = chunk_size // 10
        text_splitter = TokenTextSplitter(chunk_size=chunk_size, chunk_overlap=chunk_overlap)
        chunks_ = text_splitter.split_text(text)

        text_splitter = RecursiveCharacterTextSplitter(
            chunk_size = chunk_overlap, separators= ["\n\n", "\n", " ", "", "."], chunk_overlap=0
        )
        new_chunks_ = []
        for i in chunks_:
          cindex = text.find(i)
          if cindex!=-1 and cindex!=0:
            d = text_splitter.split_text(text[:cindex])[-1]+'\n'
            d+=i
            new_chunks_.append(d)

          else:
            new_chunks_.append(i)

        print (new_chunks_)
        db = Chroma.from_documents([Document(page_content=i) for i in new_chunks_], embeddings,
                                   persist_directory=persist_directory,
                                   client=chroma_client,
                                   client_settings=CHROMA_SETTINGS)
        print ('File has been ingested!')
        return "File has been uploaded!"
    except Exception as e:
        return e

def chat_qa(query, chat_history):
    # Use stored embeddings
    db = Chroma(persist_directory=persist_directory,
                embedding_function=embeddings,
                client_settings=CHROMA_SETTINGS,
                client=chroma_client)

    tokenizer = LlamaTokenizerFast.from_pretrained("hf-internal-testing/llama-tokenizer")
    docs = db.similarity_search(query, k=4096//300) # Get relevant docs
    print (docs)

    # Initialize chat memory, uses chat state from frontend
    memory = ConversationBufferMemory(memory_key="chat_history", input_key="question")
    chat_history_ = ''
    for index in range(0, len(chat_history), 2):
        prev_user_msg, prev_ai_msg = chat_history[index], chat_history[index+1]
        chat_history_ += "\nHuman: " + prev_user_msg
        memory.chat_memory.add_user_message(prev_user_msg)
        chat_history_ += "\AI: " + prev_ai_msg
        memory.chat_memory.add_ai_message(prev_ai_msg)

    # Selecting relevant docs
    allowed_limit = 4096-100 # 100 for safety
    allowed = False
    selected_doc = len(docs)

    while not allowed:
        print(allowed)
        context_ = ""
        for i in docs[:selected_doc]:
            context_+=i.page_content + "\n"

        prepcomputed_prompt = CHAT_PROMPT_TEMPLATE.format(chat_history=chat_history_,context=context_,question=query)
        pre_count = len(tokenizer.encode(prepcomputed_prompt))
        print(pre_count)
        if pre_count<allowed_limit:
            allowed = True
            break

        else:
            selected_doc-=1

    qa = load_qa_chain(llm=llm, memory=memory, prompt=qa_prompt, verbose=True) # Initialize QA

    result = qa({
        "input_documents" : docs[:selected_doc],
        "question" : query
    })

    return result['output_text']
```

```python
from fastapi import FastAPI, File, Request, UploadFile, Response
from fastapi.middleware.cors import CORSMiddleware
from pydantic import BaseModel

app = FastAPI()

# middlewares
app.add_middleware(
    CORSMiddleware,
    allow_origins=['*'],
    allow_credentials=True,
    allow_methods=['*'],
    allow_headers=['*'],
)

@app.get("/")
async def root():
    return {"message": "Hello World"}

@app.post("/summary")
async def PostSummary(request: Request):
    result = await request.json()
    text = result['text']
    words = result['words']
    print ('Started summarizing!')
    result = summarize_pdf(text, words)
    print ('[*] Sent summary!')
    return result

@app.post("/grammar")
async def PostGrammar(request: Request):
    result = await request.json()
    text = result['text']
    print ('grammar', text)
    result = grammar_check(text)
    print ('ans', result)
    print ('[*] Sent corrected text!')
    return result

@app.post("/ingest")
async def PostIngest(request: Request):
    result = await request.json()
    text = result['text']
    result = ingest_file(text)
    print ('[*] Uploaded file!')
    return result

@app.post("/chat")
async def PostChat(request: Request):
    result = await request.json()
    text = result['text']
    chat_history = result['chat_history']
    result = chat_qa(text, chat_history)
    print ('[*] Sent reply!')
    return result
```
